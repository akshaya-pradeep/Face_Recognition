# Attendence System

```python
import os
import cv2
import numpy as np
import pandas as pd
import dlib
import face_recognition as fr
from datetime import datetime
```

```python
path = 'Images'
```

```python
my_list = os.listdir(path)
my_list
```

```python
length = len(my_list)
length
```

```python
imgs = []
class_names = []

for img in my_list:
    imagepath = os.path.join(path,img)

    current_img = cv2.imread(imagepath)

    imgs.append(current_img)

    class_names.append(img.split('.')[0])
print(class_names)
```

```python
img = imgs[0]
conv_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_in_frame = fr.face_locations(conv_img)
print(face_in_frame)
```

```python
encode = fr.face_encodings(img,face_in_frame)[0]
print(encode)
```

```python
def face_encoding(images):
    encoded_list = []
    for img in images:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        encode = fr.face_encodings(img,face_in_frame)[0]
        encoded_list.append(encode)
    return encoded_list
```

```python
print(face_encoding(imgs))
```

```python
encoded_list = face_encoding(imgs)
encoded_list
```

```python
video = cv2.VideoCapture(0)
# Dictionary to store attendance
attendance = {}
while True:
    suc,frame = video.read()
    frame = cv2.flip(frame,1)
    frame1 = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)

    face_location = fr.face_locations(frame1)
    encoded_face = fr.face_encodings(frame1,face_location)

    for encoded_face, face_loc in zip(encoded_face, face_location):
        matches = fr.compare_faces(encoded_list,encoded_face)
        face_distances = fr.face_distance(encoded_list,encoded_face)
        match_index = np.argmin(face_distances)

        if matches[match_index]:
            name = class_names[match_index]
            if name not in attendance:
                timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                attendance[name] = timestamp
                print(f"Attendance Marked: {name} at {timestamp}")

        else:
            name = "unknown"

        y1,x2,y2,x1 = face_loc # It is the coordinate format here, not in the usual format x1,x2,y1,y2
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)

        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(45,12,122),2)

    cv2.imshow("Face",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

df = pd.DataFrame(list(attendance.items()), columns= ["Name", "Attendence Marked Time"])
df.to_csv("Attendence.csv",index=False)
print("Attendence Marked")

video.release()
cv2.destroyAllWindows()
```
