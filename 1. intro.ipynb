```python
import os 
import cv2
import numpy as np
import face_recognition as fr
```

```python
import dlib
```

```python
path = 'Images'  # path of all images will be in variable path
```

```python
# To list the images
my_list = os.listdir(path)
my_list
```

```python
length = len(my_list)
length
```

```python
# its a classification problem

imgs = []                   # To collect pixel values of input image. To assign input
class_names = []            # Corrresponding class names are in this. To assign output
for img in my_list:
    imagepath = os.path.join(path,img)
    # print(imagepath)

    current_img = cv2.imread(imagepath)
    # print(current_img)

    imgs.append(current_img)

    # To append the image name to output array calss_name we can use slicing if all the images are of same format[we need to remove the extension]
    # class_names.append(img[:,-4])

    # if all are of different format then first split and take the zeroth index
    class_names.append(img.split('.')[0])
print(class_names)

```

```python
# There will be 4 coordinates
# Face Location and Encoding (Normalisation)
# for face recognition the numpy supports only the lowest version (2+ version wont work) thats y first error occured.

img = imgs[0]
conv_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_in_frame = fr.face_locations(conv_img)
print(face_in_frame) # face location
```

```python
# Encoding the face location : To collect accurate result
encode = fr.face_encodings(img,face_in_frame)
print(encode)
```

```python
# The values are generated in many arrays we only need first array so have to take 0th index
encode = fr.face_encodings(img,face_in_frame)[0]
print(encode)
```

```python
# Function for collecting the list of images
def face_encoding(images):
    encoded_list = []
    for img in images:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame = fr.face_locations(img)
        encode = fr.face_encodings(img,face_in_frame)[0]
        encoded_list.append(encode)
    return encoded_list
```

```python
print(face_encoding(imgs))  # imgs input list
```

```python
# Assign the encoded values to a variable
encodelist = face_encoding(imgs)
print(encodelist)
```

```python
# Collecting Live video ie webcam 
video = cv2.VideoCapture(0)
while True:
    suc,frame = video.read()
    frame = cv2.flip(frame,1) 
    frame1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face_in_frame = fr.face_locations(frame1)
    # print(face_in_frame)

    encoded_face = fr.face_encodings(frame1,face_in_frame) # no need to pass [0] beacuse at a time multiple faces need to be detected when shown
    # print(encoded_face)
    
    for enc_face, face_loc in zip(encoded_face,face_in_frame):
        matches = fr.compare_faces(encodelist,enc_face)
        # print(matches)

        face_dr = fr.face_distance(encodelist,enc_face)
        # print(face_dr)

        matchindex = np.argmin(face_dr)
        # print(matchindex)

        if matches[matchindex]: # if matches and matchindex is true
            name = class_names[matchindex]
            print(name)
        else:
            name = "Unknown"

        y1,x2,y2,x1 = face_loc # It is the coordinate format here, not in the usual format x1,x2,y1,y2
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)

        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(45,12,122),2)

    cv2.imshow("Face",frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()

# [(326, 435, 378, 383)]
# []  => empty means no face is detected

# [(362, 472, 406, 429), (284, 324, 469, 139)] => two array means two faces are detected
# [(329, 482, 372, 439), (263, 324, 449, 139)]

# when i show the image of vijay then last value is true that is the corresponding index of vijay 
# print(matches)
# [False, False, False, False, True]
# [False, False, False, False, True]
# [False, False, False, False, True]
# [False, False, False, True, True]
# [False, False, False, False, True]
# [False, False, False, False, True]
# [False, False, False, False, True]
# [False, False, False, False, True]
```
